```python
data.rename(columns={u'Ваш пол':'gender',\
                     u'Как Вы приобрели билет на Ваше последнее посещение Большого Театра?':'where_got_ticket',\
                     u'Что побудило Вас посетить Большой театр?':'motive',\
                     u'Оцените от 1 до 5 Ваши впечатления от визита в Большой Театр (от 1 до 5)':'impression',\
                     u'Оцените, насколько удобен процесс покупки билетов на сайте Большого театра (от 1 до 5)':'comfort_in_buying',\
                     u'Оцените, насколько доступно и понятно на сайте Большого театра разъяснён процесс и правила продажи билетов, включая доступные льготы, график продажи, требования к документам и пр. (от 1 до 5)':'site',\
                     u'Знаете ли Вы о существовании фирменного сувенирного магазина Большого Театра?':'souvenir_shop',\
                     u'Вам было бы интересно послушать короткую бесплатную лекцию об опере/балете в здании театра перед началом представления?':'lecture',\
                     u'Оцените Ваши впечатления по шкале от 1 до 5 по следующим измерениям: [Произведение (сюжет, музыкальная основа)]':'performance',\
                     u'Оцените Ваши впечатления по шкале от 1 до 5 по следующим измерениям: [Работа оркестра]':'orchestra',\
                     u'Оцените Ваши впечатления по шкале от 1 до 5 по следующим измерениям: [Работа оперных певцов и артистов балета]':'dancers',\
                     u'Оцените Ваши впечатления по шкале от 1 до 5 по следующим измерениям: [Декорации и костюмы]':'decorations',\
                     u'Оцените Ваши впечатления по шкале от 1 до 5 по следующим измерениям: [Режиссерская интерпретация]':'producer',\
                     u'Оцените Ваши впечатления по шкале от 1 до 5 по следующим измерениям: [Буфет]':'buffet',\
                     u'Оцените Ваши впечатления по шкале от 1 до 5 по следующим измерениям: [Гардероб]':'cloakroom',\
                     u'Оцените Ваши впечатления по шкале от 1 до 5 по следующим измерениям: [Интерьер театра]':'interior',\
                     u'Как часто Вы посещаете: [Большой Театр]':'target'}, inplace = True)
```

```python
data.head()
```

```python
categorical_features = ['gender','where_got_ticket','motive','souvenir_shop','lecture']
numeric_features = list(set(data.columns)-set(categorical)-set(['target']))
```

```python
data.target = data.target.map({u'Раз в год или реже': -1.0,\
                 u'Был (а) когда-то': -1.0,\
                 u'1-2 раза в полгода': 1.0,\
                 u'1-2 раза в квартал': 1.0,\
                 u'1-2 раза в месяц': 1.0,\
                 u'Раз в неделю и чаще': 1.0})
```

```python
data.target.fillna(-1.0,inplace=True)  # u'Был (а) когда-то' не обрабатывается
```

```python
data.head(10)
```

```python
data.buffet = data.buffet.map({u'Не могу оценить': 0, 2:2, 3:3, 4:4, 5:5})
```

```python
data = pd.get_dummies(data,columns=categorical_features)
```

```python
X = data.drop('target',axis=1)
y = data.target
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,shuffle=True,random_state=0)
```

```python
param_grid = {'C': [0.01, 0.05, 0.1, 0.5, 1, 5, 10], 'penalty': ['l1','l2']}
optimizer = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='roc_auc')
```

```python
optimizer.fit(X_train,y_train)
```

```python
optimizer.best_estimator_
```

```python
from sklearn.metrics import accuracy_score
print accuracy_score(optimizer.best_estimator_.predict(X_test),y_test)
```
